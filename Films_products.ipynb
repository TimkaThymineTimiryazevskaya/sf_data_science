```python
ratings = pd.read_csv('ratings_movies.csv', sep = ',')
ratings.head(10)
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings['year_release'] = ratings['title'].apply(get_year_release)
ratings.info()
```

```python
ratings.sort_values(by =['year_release'], ascending =[False])
a = ratings.groupby(by=['title']).mean()
a[a['year_release']==1999].sort_values(by=['rating'])
```

```python
ratings.sort_values(by = ['year_release'])
a = ratings[ratings['year_release']==2010].groupby(by=['genres'])['rating'].mean()
a.sort_values()
```

```python
ratings.groupby(by=['userId'])['genres'].agg('nunique').sort_values()
```

```python
ratings.groupby(by=['userId'])['rating'].agg(['count','mean']).sort_values(by =['count', 'mean'], ascending=[True, False])
```

```python
ratings[ratings['year_release']==2018].groupby(by=['genres'])['rating'].agg(['count','mean']).sort_values(by =['count', 'mean'], ascending=[True, False])
```

```python
order = pd.read_csv('orders.csv', sep =';')
order.info()
order.head(10)
```

```python
products = pd.read_csv('products.csv', sep =';')
products.info()
products.head(10)
```

```python
order.merge(products, how ='outer', left_on = 'ID товара', right_on = 'Product_ID')
display(order)
products[products['Product_ID']==124]
```

```python
orders_products = order.merge(
    products, 
    left_on='ID товара',
    right_on='Product_ID',
    how='left')
orders_products.tail(1)['Order ID']
display(orders_products)
```

```python
orders_products[orders_products['Отменен'] == 'Да']['Name']
```

```python
orders_products['Profit'] = orders_products['Price'] * orders_products['Количество'] 
orders_products[orders_products['Оплачен'] == 'Да'].groupby('ID Покупателя')['Profit'].sum().sort_values(ascending=False)
```
