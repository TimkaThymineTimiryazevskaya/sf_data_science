```python
N = 1000 
k_ = 1 
seed = "АДВИ_2023"
random.seed(seed)

col_1 = []
for i in range(1,1001):
  col_1.append(i)
random.shuffle(col_1)
print(col_1)
 
grp_lst = ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"] 
col_2 = [] 
for i in range(N): 
    col_2.append(random.choices(grp_lst, k=k_)[0]) 
print(col_2)   
 
col_3 = [] 
for i in range(N):  
    col_3.append(random.randint(50, 100))
print(col_3)
```

```python
col_4, col_5, col_6 = [], [], [] 
mu1, mu2, mu3 = 0, 0, 50 
sigma1, sigma2, sigma3 = 1, 12, 9 
for i in range(1, N+1): 
    col_4.append(random.gauss(mu1, sigma1)) 
for i in range(1, N+1): 
    col_5.append(random.gauss(mu2, sigma2)) 
for i in range(1, N+1): 
    col_6.append(random.gauss(mu3, sigma3)) 
 
col_4[0], col_5[0], col_6[0]
```

```python
data = pd.DataFrame({'index':col_1, 'groups':col_2, 'uniform':col_3, 'gauss_1':col_4, 'gauss_2':col_5, 'gauss_3':col_6})
data
```

```python
data.index = col_1
data.info()
```

```python
df = data.copy()
```

```python
df.loc[df.index %121==0, 'gauss_1'] = None
df.loc[np.modf(df.gauss_2)[0]>0.95, 'gauss_2'] = None
df['gauss_2'].fillna(df.gauss_2.mean(), inplace = True)
df.dropna(subset = 'gauss_1', inplace = True)
```
